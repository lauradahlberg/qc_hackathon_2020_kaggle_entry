```python
X = df.drop('target', axis=1).copy()
y = df['target'].copy()

X_train, X_test, y_train, y_test, weights_train, weights_test = train_test_split(X, y, weights, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# clean the data
lr_pipe = Pipeline([('imputer', SimpleImputer(missing_values=np.nan, strategy='mean')), 
                 ('scaler', StandardScaler()), 
                 ('lr', LinearRegression())])

lr_pipe.fit(X_train, y_train)
wmse(y_test,lr_pipe.predict(X_test), weights_test)
#print("Test score: {:.2f}".format(lr_pipe.score(X_test, y_test )))
```

```python
##### clean the data
rid_pipe = Pipeline([('imputer', SimpleImputer(missing_values=np.nan, strategy='mean')), 
                 ('scaler', StandardScaler()), 
                 ('rid', Ridge(alpha = 1))])

rid_pipe.fit(X_train, y_train)
wmse(y_test,rid_pipe.predict(X_test), weights_test)
#print("Test score: {:.2f}".format(rid_pipe.score(X_test, y_test )))
```

## iterating through to find best cluster count

```python
# clean the data
score_list = []
for n in range(3,20):
    knn_pipe = Pipeline([('imputer', SimpleImputer(missing_values=np.nan, strategy='mean')), 
                 ('scaler', StandardScaler()), 
                 ('knn', KNeighborsRegressor(n_neighbors=n))])

    knn_pipe.fit(X_train, y_train)
    score_list.append({n:wmse(y_test,knn_pipe.predict(X_test), weights_test)})
#print("Test score: {:.2f}".format(knn_pipe.score(X_test, y_test )))
```

```python
score_list
```

```python
# clean the data
svr_pipe = Pipeline([('imputer', SimpleImputer(missing_values=np.nan, strategy='mean')), 
                 ('scaler', StandardScaler()), 
                 ('svr', SVR())])

svr_pipe.fit(X_train, y_train)
print({n:wmse(y_test,svr_pipe.predict(X_test), weights_test)})
#print("Test score: {:.2f}".format(knn_pipe.score(X_test, y_test )))
```

```python
# clean the data
score_list = []
for n in range(100,11000,100):
    rnn_pipe = Pipeline([('imputer', SimpleImputer(missing_values=np.nan, strategy='mean')), 
                 ('scaler', StandardScaler()), 
                 ('rnn', RadiusNeighborsRegressor(radius=n))])

    rnn_pipe.fit(X_train, y_train)
    score_list.append({n:wmse(y_test,rnn_pipe.predict(X_test), weights_test)})
#print("Test score: {:.2f}".format(knn_pipe.score(X_test, y_test )))
```

```python
score_list
```

```python
# clean the data
# clean the data
score_list = []
for n in range(10,110,10):
    pipe = Pipeline([('imputer', SimpleImputer(missing_values=np.nan, strategy='mean')), 
                 ('scaler', StandardScaler()), 
                 ('rforest', RandomForestRegressor(n_estimators = n,random_state=31))])

    pipe.fit(X_train, y_train)
    score_list.append({n:wmse(y_test,pipe.predict(X_test), weights_test)})

#print("Test score: {:.2f}".format(pipe.score(X_test, y_test )))
```

```python
score_list
```

```python
trans = Pipeline([('imputer', SimpleImputer(missing_values=np.nan, strategy='mean')),
                      ('scaler', StandardScaler())])
trans.fit(X_train)
trans_train = trans.transform(X_train)
trans_data = trans.transform(X_test)
trans_data.shape
vote3 = VotingRegressor([#('rid', Ridge(alpha = 1)), 
                         ('knn', KNeighborsRegressor(n_neighbors=6)), 
                         ('rf', RandomForestRegressor(random_state=42))])
vote3.fit(trans_train, y_train)
```

```python
preds_vote3 = vote3.predict(trans_data)
wmse(y_test,preds_vote3, weights_test)
```
